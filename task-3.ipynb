## EXPLORATORY DATA ANALYSIS

### DATA VISULISATION

```python
#visualising the dataset usingpair plot
import seaborn as sns
sns.pairplot(df)
```

```python
#finding the partwise cooreleation between the columns and visualizing using heatmaps
df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.title("corelation between varibles")
plt.show()
```

## plotting based upon the region and the category

```python
plt.figure(figsize=(6,6))
plt.title('Ship Mode')
plt.pie(df['Ship Mode'].value_counts(), labels=df['Ship Mode'].value_counts().index,autopct='%1.1f%%')
plt.show() 
```

```python
#computing top categories in terms of sales from of sales from first 100 observation
top_category_s = df.groupby("Category").Sales.sum().nlargest(n=100)
#computing top categories in terms of profit from of sales from first 100 observation
top_category_p = df.groupby("Category").Profit.sum().nlargest(n=100)


#plotting to see it vistually
plt.style.use('seaborn')
top_category_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14) 
top_category_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='red')
plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total sales/Profits',fontsize = 15)
plt.title("Top Category Sale VS Profit",fontsize = 15)
plt.show()
```

## Visualizing the Sales vs Profits in Different Sub-Categories 

```python
plt.figure(figsize = (8,8))
textprops = {"fontsize":15}
plt.title('Sub_Category')
plt.pie(df['Sub-Category'].value_counts(), labels=df['Sub-Category'].value_counts().index,textprops = textprops)
plt.show()
```

```python
#computing top categories in terms of sales from of sales from first 100 observation
top_subcategory_s = df.groupby("Sub-Category").Sales.sum().nlargest(n=100)
#computing top categories in terms of profit from of sales from first 100 observation
top_subcategory_p = df.groupby("Sub-Category").Profit.sum().nlargest(n=100)


#plotting to see it vistually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14) 
top_subcategory_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='red')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sale VS Profit",fontsize = 15)
plt.show()
```

```python
#A detail view
plt.figure(figsize=(14,12))
statewise = df.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

## Visualizing profit in different region

```python
plt.figure(figsize = (6,6))
plt.title('Region')
plt.pie(df['Region'].value_counts(), labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
#computing top region in terms of sale fromm first 100 observation
top_region_s = df.groupby("Region").Sales.sum().nlargest(n=100)
#computing top region in terms of sale fromm first 100 observation
top_region_p = df.groupby("Region").Profit.sum().nlargest(n=100)


#plotting to see it visually 
plt.style.use('seaborn')
top_region_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14) 
top_region_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='red')
plt.xlabel('Region',fontsize = 15)
plt.ylabel('Total sales/Profits',fontsize = 15)
plt.title("Top Region Sale VS Profit",fontsize = 15)
plt.show()
```

### The graph shows that west and easthave some profit through sales in the east are less as compared to west

## Visualizing the sale  vs profits in different state

```python
#computing top categories in terms of sales from of sales from first 10 observation
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)
#computing top categories in terms of profit from of sales from first 10 observation
top_states_p = df.groupby("State").Profit.sum().nlargest(n=10)


#plotting to see it vistually
plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14) 
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='red')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 States salesvs Profit",fontsize = 15)
plt.show()
```

```python
#A detail view
plt.figure(figsize=(14,12))
statewise = df.groupby(['State'])['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

### the above graph is clearly shows that California and new york have the highest sales and profit 

# checking for the interdependency of sales,profits and discounts

```python
plt.style.use('seaborn')
df.plot(kind="scatter",figsize =(10,5), x = "Sales",y="Profit",c="Discount",s=20,fontsize=16,colormap="viridis")
plt.ylabel('Total Profits,fontsize=16')
plt.title("Interdependency of sales ,Profits and Discounts",fontsize=16)
plt.show()
```

### the graph clearly shows that if we give more discounts then sales increases but profit decrease
